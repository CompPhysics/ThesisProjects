#A simplified approach to the CC amplitude-storage

The CC-amplitudes are tensors of rank 2,4,6 or possibly 8, that cause some complications when writing CC-algorithms due to their memory consumption. The most natural implementational approach is to create multidimensional arrays in which the amplitudes are called by their indices.

In this notebook we will explore an alternative implementation, in which the amplitudes are stored in list-like objects as in python. The proposed benefits from this scheme may follow from these considerations:

(1) The amplitudes will mostly contain zeros, thus it may be possible to reduce the memory cost in storage.

(2) When calculating the next iteration of amplitudes, all amplitudic products containing zeros will be zero, so a possibility of reducing the number of calculations is present.

#General description of the standard algorithm

We begin by outlining a very standard implementation of the CC algorithm, where the amplitudes are stored in arrays.

```python
Ns = 100 #number of states
Np = 14  #number of particles

t1amp = zeros((Ns,Ns))
t2amp = zeros((Ns,Ns,Ns,Ns))
#t3amp = zeros((Ns,Ns,Ns,Ns,Ns,Ns))
#t4amp = zeros((Ns,Ns,Ns,Ns,Ns,Ns,Ns,Ns))

t1amp_new = zeros((Ns,Ns))  #Next iteration
t2amp_new = zeros((Ns,Ns,Ns,Ns))
#t3amp = zeros((Ns,Ns,Ns,Ns,Ns,Ns))
#t4amp = zeros((Ns,Ns,Ns,Ns,Ns,Ns,Ns,Ns))

def v(p,q,r,s):
    #The particle-particle interaction
    pp_interaction = 0
    if p==q and r==s:
        if q==r:
            pp_interaction += 1
        if q==s:
            pp_interaction -= 1
    if p==r and q==s:
        if p==q:
            pp_interaction *= 2
        if p == s:
            pp_interaction *=3
    pp_interaction += sqrt(13.0)**2
    pp_interaction += 13**5.0
    return pp_interaction

def h(p,q):
    #The single particle energy
    h_energy = p*(p==q)
    return h_energy
```

Already, the above code will hit a wall pretty fast if we include t3amps and t4amps, and increase the number of states. 

There are basically two ways we need to interact with the stored amplitudes:

(1) Storing/updating new objects (that differ from 0).

(2) Fetching the amplitudes in the CC-summations when doing the above.

The brute force approach in CCSD (not using any symmetries) may be (mock-)written as below:

```python
#Amplitude updating:
def CC_t1amps(p,q):
    #A lot of sums of the below type, returning the next amplitude
    cct1 = 0
    for i in range(Np):
        for a in range(Np,Ns):
            cct1 += h(i,a)*t1amp[a,i]
    
    return cct1

def CC_t2amps(p,q,r,s):
    cct2 = 0
    for i in range(Np):
        for j in range(Np):
            for a in range(Np,Ns):
                for b in range(Np,Ns):
                    cct2 += v(i,j,a,b)*t2amp[a,b,i,j]
    return cct2
        
#initialization as 1.order perturbation
for a in range(Np, Ns):
    for b in range(Np, Ns):
        for i in range(Np):
            for j in range(Np):
                t2amp[a,b,i,j] = v(i,j,a,b)/(h(i,i) + h(j,j) - h(a,a) - h(b,b))

    
#Iterative step
def advance():
    for p in range(Ns):
        for q in range(Ns):
            t1amp_new[p,q] = CC_t1amps(p,q)
            for r in range(Ns):
                for s in range(Ns):
                    t2amp_new[p,q,r,s] = CC_t2amps(p,q,r,s)
    t1amp = t1amp_new
    t2amp = t2amp_new

def energy():
    #return the correlation energy
    e0 = 0.0
    for p in range(Ns):
        for q in range(Ns):
            e0 += t1amp(p,q)*h(p,q)
            for r in range(Ns):
                for s in range(Ns):
                    e0 += .5*v(p,q,r,s)*t2amp[p,q,r,s]
                    e0 += .25*v(p,q,r,s)*t1amp[p,q]*t1amp[r,s]
    return e0

#Main loop
def main_loop():
    #CCSD-iterations
    e0 = 0.0
    e_prev = 1.0
    tolerance = 10E-7
    while(abs(e0-e_prev)>tolerance):
        advance()
        e_prev = e0
        e0 = energy()
    return e0
                    
                    
```

In the above approach, the amplitudes are stored in multidimentional arrays, meaning that a certain index (p,q,r,s) corresponds to the amplitude of that element. The above code examplifies the introductory remarks;

(1) At initialization, only a fraction of the elements in the above amplitudes will be non-zero. 

(2) A lot of the products will be with amplitudes equal to zero.

To demonstrate this, we do some mock-initializations of the t2-amplitude below

```python
def ampl_init(Ns, Np):
    N_t2amps = Ns**4
    N_nonzero = 0
    for i in range(Np):
        for j in range(Np):
            for a in range(Np,Ns):
                for b in range(Np,Ns):
                    N_nonzero += 1
    return N_nonzero/float(N_t2amps)

#or simply
def amp_init(Ns,Np):
    return ((Np**2)*(Ns-Np)**2)/float(Ns**4)

Nstates = array([54,114,186,294,358])
Nonzero = []
for i in range(len(Nstates)):
    Nonzero.append(amp_init(Nstates[i], 14))
                   
#Nonzero = amp_init(Nstates, 14) #typical electron-gas setup
print Nonzero
#print amp_init(358,14)
from matplotlib.pyplot import *


figure(1)
plot(Nstates, Nonzero)
title("Fraction of non-zero elements")
xlabel("Number of states")
ylabel("Fraction of non-zero amplitudes")
show()
                
```

The figure above shows that only a very small fraction of the initialized t2-amplitudes is non-zero, and that the fraction is decreasing as the number of states is increased. It could therefore be fruitful to investigate an alternate storage routine, where only non-zero elements are stored.

#Does the order of summations matter?

When storing only non-zero elements in the amplitude arrays, we need to find a solution to how we efficiently may access a given matrix element. For example, the introduction of a sparse tensor class would make coulomn-indexing just as fast as any array-fetching, but searching the rows for a given matrix element would be significantly slower, as such algorithms must "search" through the elements in the row for a given index. In worst case-scenarios this could make access time as slow as 100ms (from some private simple experiments in python), making the lookup routine too slow for any efficient implementation of the CC-algorithm.

It may however turn out that no such searching is required, and the key may be the simple fact that the order of the summations is irrelevant.

As we have seen, the CC-algo depends on a lot of operations similar to

$$ t^{ab}_{ij} \leftarrow \sum_{klc} t^{ac}_{kl}v(a,b,i,k)$$

(This is just a mock sum, not necessarily occuring in the amplitude equations)

In the above expression we sum over three indices in the t2-amplitude, meaning that we have prior knowledge of which elements that will be accessed. If we store the amplitudes in objects where only non-zero elements and their indices is stored, we may possibly reduce memory consumption, and at the same time significantly optimize the summations (as only non-zero products would be accessed and calculated NB! Both for the pp-interaction and the amplitude access).

#Outline of a "non-zeros only" implementation

Based on the principles discussed so far, we now outline a scheme for efficient storage and calculation of amplitudes using objects similar to the pythonic lists. We divide the scheme into 3 main levels; initialization, amplitude updating and amplitude calculation. (where the last is rather a lower level of the second last). Chronologically, these will occur as follows:

1. Initialization
2. Update all amplitudes:
   3. calculate each amplitude

(3. repeat from 2 until convergence.)

The main difference from the "standard" implementation described earlier in this document is the way we iterate through the amplitude elements.

To investigate the challenges this scheme pose, we  now try to work through each step listed above by writing a mock-implementation in python.

```python
class amplitude():
    def __init__(self):
        self.elements = []
        self.N = 0 #length of list
        self.p1sort = []
        self.p2sort = []
        self.h1sort = []
        self.h2sort = []

        self.p1V = []
        self.p2V = []
        self.h1V = []
        self.h2V = []
    def at(self, i):
        #return the ith element
        return self.elements[i]
    def add(self, element):
        #append element to end of list
        self.elements.append(element)
        #if element[0] == self.p1N:
        #    self.p1sort.append(self.N)

        self.N += 1
    def map_tensor(self, maxindex):
        #Map out sequences of amplitudes corresponding to fixed indices
        self.p1N = 0
        self.p2N = 0
        self.h1N = 0
        self.h2N = 0
        for m in range(maxindex):
            self.p1V.append(self.p1N)
            self.p2V.append(self.p2N)
            self.h1V.append(self.h1N)
            self.h2V.append(self.h2N)
            
            for n in range(self.N):
                p1,p2,h1,h2, val = self.at(n)
                if p1 == m:
                    self.p1sort.append(n)
                    self.p1N += 1
                if p2 == m:
                    self.p2sort.append(n)
                    self.p2N += 1
                if h1 == m:
                    self.h1sort.append(n)
                    self.h1N += 1
                if h2 == m:
                    self.h2sort.append(n)
                    self.h2N += 1
        print len(self.h1V)
        print len(self.h1sort)
    def getlist_p1(self,n):
        #return self.elements[self.p1sort[self.p1V[n]:self.p1V[n+1]]]
        return self.p1sort[self.p1V[n]:self.p1V[n+1]]
    def getlist_p2(self,n):
        #return self.elements[self.p1sort[self.p1V[n]:self.p1V[n+1]]]
        return self.p2sort[self.p2V[n]:self.p2V[n+1]]
    def getlist_h1(self,n):
        #return self.elements[self.p1sort[self.p1V[n]:self.p1V[n+1]]]
        #print self.h1sort
        return self.h1sort[self.h1V[n]:self.h1V[n+1]]
    def getlist_h2(self,n):
        #return self.elements[self.p1sort[self.p1V[n]:self.p1V[n+1]]]
        return self.h2sort[self.h2V[n]:self.h2V[n+1]]
                
        

class amplitude2():
    def __init__(self, Nlen):
        self.indices = zeros((Nlen, 5), dtype = float)
        #self.elements = zeros((Nlen), dtype =float)
        self.Nlen = Nlen
        self.N = 0 #number of occupied elements
    def add(self, e):
        self.indices[self.N] = e
        #self.elements[self.N] = e[4]
        self.N+= 1
        if self.N>self.Nlen:
            #expand array
            return 0
    def at(self, i):
        #a = self.indices[i]
        return self.indices[i]
    
```

#2. Updating the amplitudes

This process is the actual iterative step of the algorithm. Based on the previously (or initial) amplitudes, we obtain a new set of amplitudes.  Symmetries is commonly used to reduce the computational cost. For the time being, we will follow the same procedure, but at a later stage we may find a workaround that makes the multiplicate storage of amplitudes superfluous. We still gain from the reduced computational cost, though.

```python
def advance(t2prev):
    t2new = amplitude()
    for i in range(Np):
        for j in range(i,Np):
            for a in range(Np,Ns):
                for b in range(a, Ns):
                    ampval = CC_t2amps(a,b,i,j)
                    if ampval >= 10E-12:
                        t2new.add([a,b,i,j,ampval])
                        t2new.add([b,a,i,j,-ampval])
                        t2new.add([b,a,j,i,ampval])
                        t2new.add([a,b,j,i,-ampval])
```

#3. Calculating the amplitudes

This may be the bottleneck and is probably the most tricky step of the calculation. For each combination of indices in the advance() function, we need to iterate through the amplitude lists in different ways. We have however some prior knowledge of the arrangement in the amplitude list, as its elements is appended in either the advance routine or the initialization procedure.

In the CCD case, it is natural to distinguish the quadratic contributions (t t) terms from the linear (t) terms. This is for example done in Shawitt-Bartlett, p. 262. While the indices a,b,i,j are "locked", we need to sum up the following contributions to the CCD T2-amplitude:

**Linear terms in the standard vs. "non-zeros only" implementation:**

```python
#Linear contributions to the CCD amplitude equation.
def CCD_linear_standard(a,b,i,j,t2amp):
    #Standard implementation
    l2a = 0.0
    l2b = 0.0
    l2c = 0.0
    ops = 0
    for c in range(Np, Ns):
        for d in range(Np, Ns):
            ops += 1
            l2a += v(a,b,c,d)*t2amp[c,d,i,j]

    for k in range(Np):
        for l in range(Np):
            ops += 1
            l2b += v(k,l,i,j)*t2amp[a,b,k,l]

    for k in range(Np):
        for c in range(Np, Ns):
            ops += 4
            l2c -= v(a,k,c,j)*t2amp[c,b,i,k];
            l2c += v(a,k,c,i)*t2amp[c,b,j,k];
            l2c += v(b,k,c,j)*t2amp[c,a,i,k];
            l2c -= v(b,k,c,i)*t2amp[c,a,j,k];
    print "Number of linear standard operations:", ops
    return .5*l2a + .5*l2b  + l2c

def CCD_linear(a,b,i,j, t2amps):
    #"Non-zeros only" implementation
    l2a = 0.0
    l2b = 0.0
    l2c = 0.0
    
    for n in range(t2amps.N):
        c,d,k,l, t2val = t2amps.at[n]
        if k == i and l == j:
            l2a += v(a,b,c,d)*t2val
        #Replaces:
        #for c in range(Np, Ns):
        #    for d in range(Np, Ns):
        #        l2a += v(a,b,c,d)*t2amp(c,d,i,j)
        if c==a and d==b:
            l2b += v(k,l,i,j)*t2val
        #Replaces:
        #for k in range(Np):
        #    for l in range(Np):
        #        l2b += v(k,l,i,j)*t2amp(a,b,k,l)
        if d==b and k==i:
            l2c -= v(a,k,c,j)*t2val
        if d==b and k==j:
            l2c += v(a,k,c,j)*t2val
        if d==a and k==i:
            l2c += v(a,k,c,j)*t2val
        if d==a and k==j:
            l2c -= v(a,k,c,j)*t2val
        
        #Replaces:

        #for k in range(Np):
        #    for c in range(Np, Ns):
        #        l2c -= v(a,k,c,j)*t2amp(c,b,i,k);
        #        l2c += v(a,k,c,i)*t2amp(c,b,j,k);
        #        l2c += v(b,k,c,j)*t2amp(c,a,i,k);
        #        l2c -= v(b,k,c,i)*t2amp(c,a,j,k);

    return .5*l2a + .5*l2b  + l2c
```

The result is that we now instead of having to perform all summations of products equal to zero, we may just store a list and do a series of IF-tests on each element. If the test succeeds we add the product to the sum, and in effect we need only add inn non-zero amplitudes.

There is room for even more optimalization:

- Prior knowledge of the order of the amplitudes may exclude certain sums if the constant index has passed.
- A preliminary IF-test may break the current iteration before performing any other tests.
- It is possible to combine the IF-tests so no multiple testing is performed. (The above implementation does not since it is meant to clarify the principles).
- Products containing multiple amplitudes will be double iterations through the lists.
- We may iterate through this list both ways, allowing for large indices to be found faster (using the same principle as above).
- Probably a lot more.

- The algorith should be easily extendable to T3-amplitudes.

Still a lot of details regarding all the sums in the CCD amplitudes should be adressed, and there is at least two major questions:

- Does this approach even work?
- Does this approach reduce the computational cost?
    - Worst case scenario is products containing multiple t3 amplitudes, but as we have seen it is possible to break the iteration already at the first amplitude. We may also exploit some prior knowledge of the order in the amplitude storage to optimize computational cost.
    
Moreover, it has become apparent that IF-testing is more CPU intensive than summations or even multiplications, so there should be possible to optimize further with this in mind.

```python
from time import clock

Ns = 54
Np = 14
t2amp = amplitude()
initialize_t2amps(Ns,Np,t2amp)
t2amp.map_tensor(Ns)
def CCD_linear(a,b,i,j, t2amps):
    #optimized routine I
    l2a = 0.0
    l2b = 0.0
    l2c = 0.0
    
    for n in range(t2amps.N):
        p1,p2,h1,h2, t2val = t2amps.at(n)
        #if p1>a and p2>b and h1>i:
        #    break
        #if n%1000 == 0:
        #    if p2>b and h1>i:
        #        break
            
        #<p1,p2/t/h1,h2> = val 
        if h1 == i and h2 == j:
            l2a += v(a,b,p1,p2)*t2val

        if p2==b:
            if p1 ==a:
                l2b += v(h1,h2,i,j)*t2val
            if h1==i:
                l2c -= v(a,h1,p2,j)*t2val
            if h1==j:
                l2c += v(a,h1,p1,j)*t2val
        if p2==a: 
            if h1==i:
                l2c += v(a,h1,p1,j)*t2val
            if h1==j:
                l2c -= v(a,h1,p1,j)*t2val
        

    return .5*l2a + .5*l2b  + l2c

def CCD_linear_optimized(a,b,i,j, t2amps):
    #optimized routine II
    l2a = 0.0
    l2b = 0.0
    l2c = 0.0
    ops = 0
    for n in t2amps.getlist_h1(i):
        p1,p2,h1,h2, t2val = t2amps.at(n)
        #if p1>a and p2>b and h1>i:
        #    break
        #if n%1000 == 0:
        #    if p2>b and h1>i:
        #        break
            
        #<p1,p2/t/h1,h2> = val 
        if h2 == j:
            l2a += v(a,b,p1,p2)*t2val
            ops += 1
    for n in t2amps.getlist_p2(b):
        p1,p2,h1,h2, t2val = t2amps.at(n)

        if p1 ==a:
            l2b += v(h1,h2,i,j)*t2val
            ops += 1
        if h1==i:
            l2c -= v(a,h1,p2,j)*t2val
            ops += 1
        if h1==j:
            l2c += v(a,h1,p1,j)*t2val
            ops += 1
    for n in t2amps.getlist_p2(a):
        p1,p2,h1,h2, t2val = t2amps.at(n)

        if h1==i:
            l2c += v(a,h1,p1,j)*t2val*(h1==i)
            ops += 1
        if h1==j:
            l2c -= v(a,h1,p1,j)*t2val*(h1==j)
            ops += 1
    print "Number of optimized operations:", ops
        

    return .5*l2a + .5*l2b  + l2c


```
